```python
from tensorflow.keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = (x_train / 255).astype(float)
x_test = (x_test / 255).astype(float)
```

```python
from hopfieldnn import choose_representative_samples, hopfield_train
for i in range(10):
    samples, classes = choose_representative_samples(x_train, y_train, 10)

    print(hopfield_train(classes, samples, x_train, y_train))
    print(hopfield_train(classes, samples, x_test, y_test))
```

```python
from feedforwardnn import feedforward_train
# run following command before training: rm ./logs/
feedforward_train(x_train, y_train, x_test, y_test, 10)
# run following command in directiory specified above to activate tensorboard view on training results: tensorboard --logdir logs/fit
```
